```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
datapath = "D:\\Dati_DSO\\FCC\\"
```

```python
df = pd.read_csv(datapath+'202312061650_LED_SERAMP.txt', sep=" ", header=None, names=["frame", "x1", "y1", "x2", "y2", "x3", "y3", "x4", "y4"])

df
```

```python
#dd = df.query('frame==100')
plt.plot(df.x1, df.y1, lw=0.004)
```

```python
#dd = df.query('frame==100')
import matplotlib as mlp
plt.hist2d(df.x1, df.y1,  bins=200, norm=mlp.colors.LogNorm())
plt.show()
```

```python
CH1 = []
for frame in df.frame.unique():
    d = df.query(f'frame=={frame}')
    CH1+=[np.max(np.abs(d.y1))]
CH1 = np.array(CH1)
```

```python


plt.hist(CH1, bins=50, density=True, histtype='step')

plt.legend()
#plt.yscale('log')
plt.xlabel('C1 amplitudes')
plt.title('no filter side')
plt.show()
```
