```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
sr=np.concatenate([np.arange(0, 20), np.arange(1180,1228)])
```

```python
sr
```

```python
BGO = pd.read_table(r'BGO.csv', skiprows=sr, sep=';', decimal=',', names=['wl', 'e'])
BSO = pd.read_table(r'BSO.csv', skiprows=sr, sep=';', decimal=',', names=['wl', 'e'])
PWO = pd.read_table(r'PWO_320_900nm.csv', skiprows=sr, sep=';', decimal=',', names=['wl', 'e'])
LYSO = pd.read_table(r'LYSO_medium.csv', skiprows=sr, sep=';', decimal=',', names=['wl', 'e'])
GAGG = pd.read_table(r'GAGG_medium_440nm.csv', skiprows=sr, sep=';', decimal=',', names=['wl', 'e'])
GAGG = pd.read_table(r'GAGG_medium_300nm.csv', skiprows=sr, sep=';', decimal=',', names=['wl', 'e'])
```

```python
BGO
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(BGO.wl, BGO.e/BGO.e.max(), label='BGO (ex: 300nm)')
plt.plot(BSO.wl, BSO.e/BSO.e.max(), label='BSO (ex: 300nm)')
plt.plot(PWO.wl, PWO.e/PWO.e.max(), label='PWO (ex: 300nm)')
plt.plot(LYSO.wl, LYSO.e/LYSO.e.max(), label='LYSO (ex: 300nm)')
plt.plot(GAGG.wl, GAGG.e/GAGG.e.max(), label='GAGG (ex: 420nm)')
plt.xlabel('Wavelenght [nm]', size=16)
plt.ylabel('Intensity [a.u.]', size=16)
plt.grid()
plt.title('FCC-NA Emission spectra meas. @ UNINA phot. lab', size=14)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=14)
plt.xticks(np.arange(300, 901, 50))
plt.savefig('emission.png', dpi=300, bbox_inches='tight')
plt.show()
```
