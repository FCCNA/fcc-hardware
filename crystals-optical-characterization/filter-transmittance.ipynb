```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
UG1 = pd.read_table(r'20240118_filter-and-crystal_transmittance/SampleUG1.Sample.asc', skiprows=90, sep='\t', names=['wl', 't'])
LP665 = pd.read_table(r'20240118_filter-and-crystal_transmittance/Sample665.Sample.asc', skiprows=90, sep='\t', names=['wl', 't'])
```

```python
UG1
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(UG1.wl, UG1.t, label='BGO')
plt.xlabel('Wavelenght [nm]', size=16)
plt.ylabel('Transmittance [%]', size=16)
plt.grid()
plt.title('UG1', size=14)
#plt.legend(fontsize=14, loc=4)
plt.xticks(np.arange(200, 900, 50))
plt.xlim(200, 900)
plt.ylim(-5, 100)
plt.show()
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(UG1.wl, UG1.t, label='BGO')
plt.xlabel('Wavelenght [nm]', size=16)
plt.ylabel('Transmittance [%]', size=16)
plt.grid()
plt.title('UG1', size=14)
#plt.legend(fontsize=14, loc=4)
plt.xticks(np.arange(200, 901, 50))
plt.xlim(200, 850)
plt.ylim(-5, 100)
plt.savefig('UG1.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(LP665.wl, LP665.t, label='BGO')
plt.xlabel('Wavelenght [nm]', size=16)
plt.ylabel('Transmittance [%]', size=16)
plt.grid()
plt.title('Longpass 665 nm', size=14)
#plt.legend(fontsize=14, loc=4)
plt.xticks(np.arange(200, 901, 50))
plt.xlim(200, 850)
plt.ylim(-5, 100)
plt.savefig('665.png', dpi=300, bbox_inches='tight')
plt.show()
```
