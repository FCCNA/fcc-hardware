```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
BGO = pd.read_table(r'BGO.Sample.csv', skiprows=2,sep=',', names=['wl', 't'])
PWO = pd.read_table(r'PWO.Sample.csv', skiprows=2,sep=',', names=['wl', 't'])
LYSO = pd.read_table(r'LYSO.Sample.csv', skiprows=2,sep=',', names=['wl', 't'])
GAGG = pd.read_table(r'GAGG.Sample.csv', skiprows=2,sep=',', names=['wl', 't'])
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(BGO.wl, BGO.t, label='BGO')
plt.plot(PWO.wl, PWO.t, label='PWO')
plt.plot(LYSO.wl, LYSO.t, label='LYSO')
plt.plot(GAGG.wl, GAGG.t, label='GAGG')
plt.xlabel('Wavelenght [nm]', size=16)
plt.ylabel('Transmittance [%]', size=16)
plt.grid()
plt.legend(fontsize=14)
plt.savefig('transmittance.png', dpi=300, bbox_inches='tight')
plt.show()
```
