```python
dframes = {'nof_h': df_nofilters,
           'ug1_v': df_UG1v,
           'ug1_h': df_UG1,
           'sp_h': df_shortpass}

amplitudes = {}
```

```python
for k in dframes.keys():
    df = dframes[k]
    CH2 = []
    CH1 = []
    for frame in df.frame.unique():
        d = df.query(f'frame=={frame}')
        CH2+=[np.max(np.abs(d.C2))]
        CH1+=[np.max(np.abs(d.C1))]
    CH1 = np.array(CH1)
    CH2 = np.array(CH2)
    amplitudes[k] = {'ch1': CH1, 'ch2': CH2 }
    print('done')
```

```python
def func(x, a, tr, tf, t0, c):
    d=0
    y = c + (a*(1-np.exp(-(x-t0)/tr))*(np.exp(-(x-t0)/tf))+d)*(x>t0) + 0*(x<=t0)
    return y

from iminuit import Minuit
from iminuit.cost import LeastSquares

wf = pd.DataFrame({'x':(hx[:-1]+hx[1:])/2, 'y':hy})

wf_cutted = wf.query('x<200E-9')
xdata = wf_cutted.x.values
ydata = wf_cutted.y.values
sigmas=1

least_squares = LeastSquares(xdata*1E9, ydata, sigmas, func)
m = Minuit(least_squares, a=1, tr=25, tf=300, t0=-400, c=0.05)#, d=-0.4)

m.params

m.migrad();

plt.scatter(xdata*1E9, ydata, s=5, color='black', label="data")
plt.plot(xdata*1E9, func(xdata*1E9, *m.values), label="fit", color='red')
fit_info = [] #[ f"$\\chi^2$ / $n_\\mathrm{{dof}}$ = {m.fval:.1f} / {len(xdata) - m.nfit}" ]
for p, v, e in zip(m.parameters, m.values, m.errors):
    fit_info.append(f"{p} = ${v:.3f} \\pm {e:.3f}$")
plt.legend(title="\n".join(fit_info))
plt.xlabel('Time [ns]')
plt.ylabel('y [V]')
plt.grid()
plt.show()
```
