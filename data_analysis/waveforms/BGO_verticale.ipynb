```python
dd = df_v.query('2<frame')
frame = dd.frame.min()
cv = dd.query(f'frame=={frame}').Time.values
binwidth = abs(cv[1]-cv[0])/2
xbins = cv - binwidth
fxbins = np.append(xbins, [xbins[-1]+binwidth])

hy, hx = np.histogram(dd.Time , weights=dd.C1/len(dd.frame.unique()), bins=fxbins)
hx = (hx[1:]+hx[:-1])/2
plt.plot(hx, hy, label = 'C1')
plt.title('BGO vertical')
plt.xlabel('Time [ns]')
plt.ylabel('Amplitude [V]')
plt.legend()
plt.grid()
plt.show()
```

```python
CH2 = []
CH1 = []
for frame in df_pwo.frame.unique():
    d = df_pwo.query(f'frame=={frame}')
    CH2+=[np.max(np.abs(d.C2))]
    CH1+=[np.max(np.abs(d.C1))]
CH1 = np.array(CH1)
CH2 = np.array(CH2)
print('done')

plt.hist(CH1, bins=50, density=True, histtype='step', label='C1')

plt.legend()
plt.yscale('log')
plt.xlabel('C1 ')
plt.title('C2')
plt.show()
```
