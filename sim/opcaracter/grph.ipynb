```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import glob
folder=glob.glob('opcara/*.asc')

files=[]

for filename in folder:
    file = pd.read_csv(filename, names=['lambd', 'pc'], skiprows=90, sep='\t', decimal='.')
    files.append(file)


plt.figure(figsize=(20, 20))
for x in files:
    plt.plot(x.lambd,x.pc)
plt.grid(True)
plt.legend(folder,loc='upper center')
plt.show()

```
