```python
import ROOT
```

```python
#prendo file simulati
simulazione = ROOT.TFile.Open("simul.root")
sc0 = simulazione.carica0
sc = simulazione.carica

vsc0 = simulazione.vcarica0
vsc = simulazione.vcarica

#prendo file misurati
esperimento = ROOT.TFile.Open("espe.root")
ec0 = esperimento.carica0
ec = esperimento.carica

vec0 = esperimento.vcarica0
vec = esperimento.vcarica
```

```python


ec.SetLineWidth(2)
ec0.SetLineWidth(2)
sc.SetLineWidth(2)
sc0.SetLineWidth(2)
ec.SetLineColor(1)
ec0.SetLineColor(1)
sc.SetLineColor(1)
sc0.SetLineColor(1)

sc0.SetLineStyle(1)
sc.SetLineStyle(1)

ec0.SetFillColorAlpha(16,0.5)
ec.SetFillColorAlpha(16,0.5)
ec0.SetLineColorAlpha(16,0)
ec.SetLineColorAlpha(16,0)

ec0.SetTitle('Esperimento')
sc0.SetTitle('Simulazione')
ec.SetTitle('Esperimento')
sc.SetTitle('Simulazione')


canvas0 = ROOT.TCanvas()
canvas0.cd()
pad0=ROOT.TPad()
pad0.Draw()
pad0.cd()
zero=ROOT.THStack("stack0", "Istogramma simulazione ed esperimento;Carica raccolta (u.a.)")
zero.Add(ec0)
zero.Add(sc0)

zero.Draw("nostack")
pad0.BuildLegend(0.80,0.85,0.99,0.95,"")
canvas0.Draw()

canvas = ROOT.TCanvas()
canvas.cd()
pad=ROOT.TPad()
pad.Draw()
pad.cd()
cut=ROOT.THStack("stack", "Istogramma simulazione ed esperimento;Carica raccolta (u.a.)")
cut.Add(ec)
cut.Add(sc)

cut.Draw("nostack")
pad.BuildLegend(0.80,0.85,0.99,0.95,"")
canvas.Draw()

```

```python
canvas0.SaveAs("Figures/name.png")
canvas.SaveAs("Figures/namecut.png")
```

```python

vec.SetLineWidth(2)
vec0.SetLineWidth(2)
vsc.SetLineWidth(2)
vsc0.SetLineWidth(2)
vec.SetLineColor(1)
vec0.SetLineColor(1)
vsc.SetLineColor(1)
vsc0.SetLineColor(1)

vsc0.SetLineStyle(1)
vsc.SetLineStyle(1)

vec0.SetFillColorAlpha(16,0.5)
vec.SetFillColorAlpha(16,0.5)
vec0.SetLineColorAlpha(16,0)
vec.SetLineColorAlpha(16,0)

vec0.SetTitle('Esperimento')
vsc0.SetTitle('Simulazione')
vec.SetTitle('Esperimento')
vsc.SetTitle('Simulazione')


vcanvas0 = ROOT.TCanvas()
vcanvas0.cd()
vpad0=ROOT.TPad()
vpad0.Draw()
vpad0.cd()
vzero=ROOT.THStack("stack0", "Istogramma simulazione ed esperimento;Carica raccolta (u.a.)")
vzero.Add(vec0)
vzero.Add(vsc0)

vzero.Draw("nostack")
vpad0.BuildLegend(0.80,0.85,0.99,0.95,"")
vcanvas0.Draw()

vcanvas = ROOT.TCanvas()
vcanvas.cd()
vpad=ROOT.TPad()
vpad.Draw()
vpad.cd()
vcut=ROOT.THStack("stack", "Istogramma simulazione ed esperimento;Carica raccolta (u.a.)")
vcut.Add(vec)
vcut.Add(vsc)

vcut.Draw("nostack")
vpad.BuildLegend(0.80,0.85,0.99,0.95,"")
vcanvas.Draw()
```
